```python
import pandas as pd
import numpy as np
from pandas import ExcelFile
import glob
```

```python
file_name=glob.glob("*.xls")
```

```python
prov_name =['Bình Thuận','Cà Mau','Lâm Đồng','Sóc Trăng','Bình Dương','Lào Cai','Bến Tre','Long An', 'Nam Định','Trà Vinh',
           'Nghệ An','Phú Thọ','Quảng Bình','Quảng Nam','Quảng Ngãi','Đà Nẵng','Hà Nội','Sơn La','Tuyên Quang','Bắc Kạn',
           'Hà Giang','Ninh Thuận','Thanh Hóa','Vĩnh Phúc','Bắc Giang','Bình Phước','Cao Bằng','Đồng Tháp','Thừa Thiên Huế',
           'An Giang','Bắc Ninh','Bình Định','Điện Biên','Lạng Sơn','Bạc Liêu','Gia Lai','Hải Dương','Hà Nam','Yên Bái',
           'Đồng Nai','Hậu Giang','Hòa Bình','Ninh Bình','Quảng Ninh','Hà Tĩnh','Hưng Yên','Kon Tum','Tiền Giang','Cần Thơ',
           'Đắk Lắk','Kiên Giang','Phú Yên','Đắk Nông','Khánh Hòa','Lai Châu','Vĩnh Long','Bà Rịa Vũng Tàu','Hải Phòng',
           'Quảng Trị','Tây Ninh','Thái Bình','Thái Nguyên','TP. Hồ Chí Minh']
```

```python
df =pd.read_excel(file_name[0])
df['Tỉnh Thành']=prov_name[0]
df
```

```python
for i in  range(1,len(file_name)):
    temp_df = pd.read_excel(file_name[i])
    temp_df['Tỉnh Thành']=prov_name[i]
    df=pd.concat([df,temp_df],ignore_index=True)
```

```python
df[10000:12000]
```

```python
df = df.drop(df.columns[[0, 1]], axis=1)
df.head()
```

```python
df['DIEM_THI']=df['DIEM_THI'].str.replace('H: ','H:   ')
df['DIEM_THI']=df['DIEM_THI'].str.replace('N: ','N:   ')
```

```python
df['DIEM_THI']=df['DIEM_THI'].str.split('   ')
```

```python
toan=[]
van=[]
li=[]
hoa=[]
sinh=[]
khtn=[]
su=[]
dia=[]
cd=[]
khxh=[]
anh=[]
```

```python
def create_value(point,name,arr):
    try:
        i = point.index(name)
        arr.append(point[i+1])
    except:
        arr.append('')
```

```python
for point in df['DIEM_THI']:
    create_value(point,'Toán:',toan)
    create_value(point,'Ngữ văn:',van)
    create_value(point,'Vật lí:',li)
    create_value(point,'Hóa học:',hoa)
    create_value(point,'Sinh học:',sinh)
    create_value(point,'KHTN:',khtn)
    create_value(point,'Lịch sử:',su)
    create_value(point,'Địa lí:',dia)
    create_value(point,'GDCD:',cd)
    create_value(point,'KHXH:',khxh)
    create_value(point,'Tiếng Anh:',anh)
```

```python
df['Toán']=toan
df['Ngữ văn']=van
df['Vật lí']=li
df['Hóa học']=hoa
df['Sinh học']=sinh
df['KHTN']=khtn
df['Lịch sử']=su
df['Địa lí']=dia
df['GDCD']=cd
df['KHXH']=khxh
df['Tiếng Anh']=anh
df.head
```
